```python
import tarfile

filename = "aclImdb_v1.tar.gz"
extract_path = "aclImdb"

with tarfile.open(filename, "r:gz") as tar:
    tar.extractall()
```

```python
import os

def load_data(foldername, sentiment):
    data = []
    for filenames in os.listdir(foldername):
        file_path = os.path.join(foldername, filenames)
        if os.path.isfile(file_path):
            with open(file_path, 'r', encoding='utf-8') as file:
                review = file.read()
                data.append((review, sentiment))
    return data
```

```python
import pandas as pd

train_pos = load_data('aclImdb/train/pos', "positive")
train_neg = load_data('aclImdb/train/neg', "negative")

train_data = train_pos + train_neg
train_df = pd.DataFrame(train_data, columns=['review', 'sentiment'])
```

```python
test_pos = load_data('aclImdb/test/pos', "positive")
test_neg = load_data('aclImdb/test/neg', "negative")
test_data = test_pos + test_neg
test_df = pd.DataFrame(test_data, columns=['review', 'sentiment'])
```

```python
train_df = train_df.sample(frac=1, random_state=42).reset_index(drop=True)
train_df.to_csv('train.csv', index=False)
```

```python
test_df = test_df.sample(frac=1, random_state=42).reset_index(drop=True)
test_df.to_csv('test.csv', index=False) 
```
